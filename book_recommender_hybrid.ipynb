```python

ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
# avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
# avg_rating_df
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the column
avg_rating_df.rename(columns={"Book-Rating": "avg_rating"}, inplace=True)

# Display the DataFrame
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
# Check if 'num_ratings' column exists before filtering and sorting
if 'num_ratings' in popular_df.columns:
    popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values(by='avg_rating', ascending=False).head(50)
else:
    print("'num_ratings' column does not exist in the DataFrame.")
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# Sample a subset of books to reduce input size
sampled_books = books.sample(n=5000, random_state=42)  # Adjust n to suit your environment
sampled_books['combined_features'] = sampled_books['Book-Author'].fillna('') + ' ' + sampled_books['Publisher'].fillna('')
# Vectorize the combined features for the sampled subset
tfidf = TfidfVectorizer(stop_words='english', max_features=1000)  # Limit to top 1000 features
tfidf_matrix = tfidf.fit_transform(sampled_books['combined_features'])

# Compute similarity matrix for the subset
similarity_matrix = cosine_similarity(tfidf_matrix)
```

# Collaborative filltering based

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:10]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

# Hybrid Recommender System

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
def collect_user_feedback(user_id, book_title, feedback):
    global user_feedback
    new_entry = pd.DataFrame({
        'User-ID': [user_id],
        'Book-Title': [book_title],
        'Feedback': [feedback]
    })

    user_feedback = pd.concat([user_feedback, new_entry], ignore_index=True)

# Example of collecting feedback
collect_user_feedback(123, "To Kill a Mockingbird", "like")
collect_user_feedback(123, "American Gods", "dislike")
```

```python
books = pd.read_pickle('books.pkl')
pt = pd.read_pickle('pt.pkl')
similarity_scores = pd.read_pickle('similarity_scores.pkl')
popular_df = pd.read_pickle('popular.pkl')
```

```python
def hybrid_recommend(user_id, book_name):
    # Collaborative filtering recommendation
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:5]

    collaborative_rec = []
    for i in similar_items:
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        collaborative_rec.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))

    # Popularity-based recommendation
    popular_rec = popular_df['Book-Title'].tolist()

    # Content-based recommendation
    try:

        book_idx = books[books['Book-Title'] == book_name].index[0]


        content_similarity_scores = list(enumerate(similarity_matrix[book_idx]))


        sorted_content_books = sorted(content_similarity_scores, key=lambda x: x[1], reverse=True)[1:5]

        # Fetch book titles
        content_rec = [books.iloc[i[0]]['Book-Title'] for i in sorted_content_books]
    except IndexError:
        content_rec = []

    # Hybridize recommendations
    hybrid_rec = list(set(collaborative_rec + popular_rec + content_rec))[:10]

    # Filter based on user feedback
    liked_books = user_feedback[(user_feedback['User-ID'] == user_id) & (user_feedback['Feedback'] == 'like')]['Book-Title'].tolist()
    disliked_books = user_feedback[(user_feedback['User-ID'] == user_id) & (user_feedback['Feedback'] == 'dislike')]['Book-Title'].tolist()

    # Prioritize liked books and exclude disliked ones
    final_recommendations = [book for book in hybrid_rec if book not in disliked_books]


    final_recommendations = list(set(liked_books + final_recommendations))[:10]  # Get top 10 unique recommendations

    return final_recommendations
```

```python
recommendations = hybrid_recommend(user_id=123, book_name="To Kill a Mockingbird")
print(recommendations)
```
